## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# function 

def generate_uniform_random_numbers(bottom, ceiling, count):
    return np.random.uniform(bottom, ceiling, count)
```

```python
#for both examples bellow

dataset1 = generate_uniform_random_numbers(10, 15, 100)
dataset2 = generate_uniform_random_numbers(10, 60, 1000)
```

```python
# Plot histograms

# Subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot histograms 
axs[0].hist(dataset1, bins=10, alpha=0.7, color='blue', label='Dataset 1')
axs[1].hist(dataset2, bins=10, alpha=0.7, color='green', label='Dataset 2')

# Lables
axs[0].set_title('Dataset 1 Histogram')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Count')
axs[0].legend()

axs[1].set_title('Dataset 2 Histogram')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Count')
axs[1].legend()

# Layout
plt.tight_layout()

# Show
plt.show();
```

How are the two distributions different?

```python
# histogram 2 is more 'uniforme' than histogram 1 (different ranges, different # of datapoints)
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
#function

def generate_normal_random_numbers(mean, std_dev, count):
    return np.random.normal(mean, std_dev, count)
```

```python
#two different scenarios

dataset1 = generate_normal_random_numbers(10, 1, 1000)
dataset2 = generate_normal_random_numbers(10, 50, 1000)
```

```python
# Plot

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Subplots
axs[0].hist(dataset1, bins=30, alpha=0.7, color='blue', label='Dataset 1')
axs[1].hist(dataset2, bins=30, alpha=0.7, color='green', label='Dataset 2')

# Labels
axs[0].set_title('Dataset 1 Histogram')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Count')
axs[0].legend()

axs[1].set_title('Dataset 2 Histogram')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Count')
axs[1].legend()

# Layout
plt.tight_layout()

# Show 
plt.show()
```

How are the two distributions different?

```python
# Dataset1: narrower distribution, with values concentrated around the mean (lower standarddev)
# Dataset 2: wider distribution, with values more spread around the mean (hihger standarddev)
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
#function

def generate_exponential_numbers(mean, size):
    return np.random.exponential(scale=mean, size=size)
```

```python
#datasets

dataset1 = generate_exponential_numbers(10, 10)
dataset2 = generate_exponential_numbers(10, 100)
```

```python
# Subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Histograms
axs[0].hist(dataset1, bins=100, alpha=0.7, color='blue', label='Size 10')
axs[1].hist(dataset2, bins=100, alpha=0.7, color='green', label='Size 100')

# Lable
axs[0].set_title('Size 10 Histogram')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Count')
axs[0].legend()

axs[1].set_title('Size 100 Histogram')
axs[1].set_xlabel('Value')
axs[1].set_ylabel('Count')
axs[1].legend()

# Layout
plt.tight_layout()

# Plot
plt.show()
```

How are the two distributions different?

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
lambda_param = 1/10
time = 15
```

```python
probability_less_15 = 1 - np.exp(-lambda_param * time)
round(probability_less_15,3)
```

What is the probability that the customer will spend more than 15 minutes

```python
1 - probability_less_15
```
